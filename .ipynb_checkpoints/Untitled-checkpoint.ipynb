```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
pd.set_option("display.max_columns", 50)
```

```python
data = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
data.head()
```

```python
encoder = LabelEncoder()
data["customerID"] = encoder.fit_transform(data["customerID"]).astype('int')
data["gender"] = encoder.fit_transform(data["gender"]).astype('int')
data['SeniorCitizen'] = data['SeniorCitizen'].astype('int')
data["Partner"] = encoder.fit_transform(data["Partner"]).astype('int')
data["Dependents"] = encoder.fit_transform(data["Dependents"]).astype('int')
data['tenure'] = data['tenure'].astype('int')
data["PhoneService"] = encoder.fit_transform(data["PhoneService"]).astype('int')
data["MultipleLines"] = encoder.fit_transform(data["MultipleLines"]).astype('int')
data["InternetService"] = encoder.fit_transform(data["InternetService"]).astype('int')
data["OnlineSecurity"] = encoder.fit_transform(data["OnlineSecurity"]).astype('int')
data["OnlineBackup"] = encoder.fit_transform(data["OnlineBackup"]).astype('int')
data["DeviceProtection"] = encoder.fit_transform(data["DeviceProtection"]).astype('int')
data["TechSupport"] = encoder.fit_transform(data["TechSupport"]).astype('int')
data["StreamingTV"] = encoder.fit_transform(data["StreamingTV"]).astype('int')
data["StreamingMovies"] = encoder.fit_transform(data["StreamingMovies"]).astype('int')
data["Contract"] = encoder.fit_transform(data["Contract"]).astype('int')
data["PaperlessBilling"] = encoder.fit_transform(data["PaperlessBilling"]).astype('int')
data["PaymentMethod"] = encoder.fit_transform(data["PaymentMethod"]).astype('int')
data['MonthlyCharges'] = data['MonthlyCharges'].astype('float64')
data['TotalCharges'] = data['TotalCharges'].convert_objects(convert_numeric=True).fillna(0)
data["Churn"] = encoder.fit_transform(data["Churn"]).astype('int')
```

```python
labels = data.Churn
features = data.drop(['customerID','Churn'],axis = 1)
```

```python
train_f, test_f, train_l, test_l = train_test_split(features, labels, test_size = 0.2, random_state = 522)
```

```python
clf = LogisticRegression(random_state = 522)
clf.fit(train_f,train_l)
predict = clf.predict(test_f)
```
